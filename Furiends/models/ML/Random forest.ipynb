```python
import os
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV

url = 'https://raw.githubusercontent.com/gggytan/Furiends/main/Furiends/data/raw/Fake%20Data2/data_with_satisfication.csv?token=GHSAT0AAAAAAB75W7AQNOFEYCLCRE6NE4DQZB652RQ'
data = pd.read_csv(url)
data
```

```python
data["category_name"] = data["category_name"].replace({"'Bulldog'": 1,
                                                      "'Maltese'": 2,
                                                      "'Poodle'": 3,
                                                      "'German Shepherd'": 4,
                                                      "'Golden Retriever'": 5,
                                                      "'Shi-ba-inu'": 6})
```

```python
X = data[["Age", "Gender", "Rehabilitation", "Pet_intimacy", "log_count", "com_count", "avg_loading", "category_name", "sum_duration"]]
Y = data[["User_satisfication"]]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# 初始化随机森林回归器
#n_estimators 设置tree的数量，使用fit拟合结果
rf = RandomForestRegressor(n_estimators=150, random_state=42)

# 拟合模型并预测结果
rf.fit(X_train, Y_train.values.ravel())
Y_pred = rf.predict(X_test)
```

```python
# 输出模型的评估指标
print("R^2 score: {:.2f}".format(rf.score(X_test, Y_test)))
```

```python
#pring mse
mse = mean_squared_error(Y_test, Y_pred)
rmse = mse**.5
print(mse)
print(rmse)
```

```python
#模型优化
# 定义超参数空间
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

# 初始化随机森林回归器和网格搜索
# GridSearchCV：Grid Search交叉验证，用于系统地遍历多种参数组合，通过交叉验证确定最佳效果参数。
# rf：指定要优化的模型，这里是随机森林模型。
# param_grid：给定要优化的超参数及其取值范围，它是一个字典类型的对象，其中key是要优化的超参数名，value是超参数的取值范围，如'n_estimators': [50, 100, 200], 'max_depth': [10, 20, 30], 'min_samples_split': [2, 5, 10]。
# cv：交叉验证的折数，这里是5。
# n_jobs：指定用于计算的CPU数量，-1表示使用所有可用的CPU
rf = RandomForestRegressor(random_state=42)
grid_search = GridSearchCV(rf, param_grid, cv=5, n_jobs=-1)

# 拟合模型并预测结果
grid_search.fit(X_train, Y_train.values.ravel())
Y_pred = grid_search.predict(X_test)

# 输出模型的评估指标和最佳超参数组合
print("R^2 score: {:.2f}".format(grid_search.score(X_test, Y_test)))
print("Best parameters: ", grid_search.best_params_)
mse = mean_squared_error(Y_test, Y_pred)
rmse = mse**.5
print(mse)
print(rmse)
```

```python
rf = RandomForestRegressor(n_estimators=grid_search.best_params_['n_estimators'],
                            max_depth=grid_search.best_params_['max_depth'],
                            min_samples_split=grid_search.best_params_['min_samples_split'],
                            random_state=42)
rf.fit(X_train, Y_train.values.ravel())
Y_pred = rf.predict(X_test)
# 获取特征重要性分数
importance_scores = rf.feature_importances_
# 将特征重要性分数按从大到小排序，并输出特征名和分数
importance_df = pd.DataFrame({'feature_name': X.columns, 'importance_score': importance_scores})
importance_df = importance_df.sort_values(by='importance_score', ascending=False)
print(importance_df)
```
