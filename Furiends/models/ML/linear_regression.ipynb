```python
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
dummy = pd.get_dummies(df["category_name"])
dummy.columns = dummy.columns.str.replace("'", "")
new = pd.concat([df,dummy],axis=1)
```

```python
new
```

```python
new.columns
```

```python
cols = ["Age", "Gender", "Rehabilitation","Pet_intimacy", "log_count", "com_count", "avg_loading", "sum_duration",'Bulldog', 'German Shepherd', 'Golden Retriever', 'Maltese', 'Poodle',
       'Shi-ba-inu']
X_train = new[cols]
X_train = sm.add_constant(X_train)
y_train = df['User_satisfication']
simple_lr = sm.OLS(y_train, X_train).fit() #ordinary least square
print(simple_lr.summary())
```

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python
#The fit_regularized() method in Statsmodels can be used to fit regularized regression models, including Lasso and Ridge regression.
lasso = sm.OLS(y_train, X_train).fit_regularized(method='elastic_net', alpha=0.1, L1_wt=1.0)
ridge = sm.OLS(y_train, X_train).fit_regularized(method='elastic_net', alpha=0.1, L1_wt=0)
```

```python
# code for finding best alpha
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
import numpy as np

ridge = Ridge()
param_grid = {'alpha': np.linspace(0, 1, 100)}
grid_search = GridSearchCV(ridge, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

best_alpha = grid_search.best_params_['alpha']
best_model = grid_search.best_estimator_
```

```python
# rank of coefficients of the factors 
lasso.params
```

```python
lasso_coefs = pd.DataFrame({'feature': range(0, 15),
                            'coef': lasso.params,
                            'abs_coef': abs(lasso.params)})
```

```python
# rank of coefficients of the factors 
lasso_coefs.sort_values('abs_coef',ascending = False)
```

```python
y_lasso_pred = lasso.predict(X_train)
y_ridge_pred = ridge.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error

mse_lasso = mean_squared_error(y_train, y_lasso_pred)
mse_ridge = mean_squared_error(y_train, y_ridge_pred)
```

```python
print("mse_lasso:" ,mse_lasso)
print("mse_ridge:",mse_ridge)
```
